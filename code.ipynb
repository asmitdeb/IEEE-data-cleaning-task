```python
import pandas as pd
sheet1 = pd.read_csv('sheet1.csv',index_col = False)
sheet2 = pd.read_csv('sheet2.csv',index_col = False)
len1 = len(sheet1)
len2 = len(sheet2)
```

```python
sheet1 = sheet1.drop(columns = ["Leader College","Leader Year"])
```

```python
sheet2 = sheet2.drop(columns = ["Team Id", "Candidate's Gender", "Country", "Course Pursuing", "Course Stream", "Specialization", "Year Of Study", "Passing Out Year", "Candidate's Organisation", "Registration Time"])
```

```python
sheet1.rename(columns = {"Leader Name" : "Candidate's Name", "Leader Email" : "Candidate's Email", "Leader Phone" : "Candidate's Mobile"}, inplace = True)
```

```python
memtype = ["team leader"]*len(sheet1)
```

```python
sheet1["Candidate Type"] = memtype
```

```python
sheet3 = pd.concat([sheet1, sheet2],ignore_index = True) #merge the 2 sheets
```

```python
sheet3["Candidate's Mobile"] = sheet3["Candidate's Mobile"]%10000000000 #format the phone numbers
```

```python
sheet3 = sheet3.drop(columns = ["Members"])
```

```python
sheet3 = sheet3.drop_duplicates() #remove duplicate entries
```

```python
sheet3.rename(columns = {"Candidate's Name":"Participant Name", "Candidate's Email":"Participant Email", "Candidate's Mobile":"Participant Phone Number", "Candidate Type":"Participant Type"},inplace = True)
```
